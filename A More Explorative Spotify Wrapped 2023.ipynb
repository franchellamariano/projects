### My Most Played Songs in Each Month
This detailed analysis not only highlights my musical preferences but also connects the songs to specific experiences and moods. Spotify Wrapped doesn't provide users with their most listened-to songs in each month, making this personalized analysis an invaluable addition to understand the ebb and flow of musical preferences throughout the year.

```python

def most_played_in_month(data, excluded_artists=None):
    data['endTime'] = pd.to_datetime(data['endTime'])
    data['msPlayed'] = data['msPlayed'] / 60000
    if excluded_artists:
        data = data[~data['artistName'].isin(excluded_artists)]
    monthly_playtime = data.groupby([data['endTime'].dt.year, data['endTime'].dt.month, 'artistName', 'trackName'])['msPlayed'].sum()
    most_played_in_month = monthly_playtime.groupby(level=[0, 1]).idxmax().apply(lambda x: x[3])
    return most_played_in_month
excluded_artists = ['The Tiny Meat Gang Podcast', 'Theo Von Funniest Moments']
result = most_played_in_month(data, excluded_artists)
print("Most played songs in each month:")
print(result)
```

**August**: August marked the peak of summer vacation, and "1999" by Prince encapsulated the spirit of having fun and carefree enjoyment. The remastered version likely intensified the nostalgia, making it a perfect anthem for my summertime.


**September**: Unsurprisingly, Earth, Wind & Fire's "September" claims the top spot in September. Beyond the thematic alignment with the month's title, the mention of "the 21st night of September" adds a serendipitous touch, making it a fitting soundtrack for the month.


**October**: "Sweat (A La La La La Long)": October embraced the reggae-infused rhythms of "Sweat (A La La La La Long)." The laid-back vibes of reggae music along with the start of my obsession with reggae during this time provided a pleasant and relaxing auditory experience during this month.

### My Most Played Artists Duirng Each Month
This thorough examination not only pinpoints my prominent artists but also underscores the significance of album releases, illustrating how major musical events shapeed my monthly listening preferences. Significantly, Spotify Wrapped also lacks the feature to reveal users' most-listened-to artists each month, underscoring the invaluable contribution of this personalized analysis in comprehending the nuanced evolution of musical preferences across time.

```python
def most_played_artists_per_month(data, excluded_artists=None):
    if excluded_artists:
        data = data[~data['artistName'].isin(excluded_artists)]
    monthly_playtime = data.groupby([data['endTime'].dt.year, data['endTime'].dt.month, 'artistName'])['msPlayed'].sum()
    most_played_artists_per_month = monthly_playtime.groupby(level=[0, 1], group_keys=False).idxmax().apply(lambda x: x[2])
    return most_played_artists_per_month
excluded_artists = ['The Tiny Meat Gang Podcast', 'Theo Von Funniest Moments']
result = most_played_artists_per_month(data, excluded_artists)
print("Most played artists in each month:")
print(result)
```

**August**: August featured the dynamic sounds of Travis Scott, with the release of his album "Utopia." The album's impact likely contributed to his dominance in my playlist, capturing the essence of the month's musical landscape.

**September**: Playboi Carti's distinct style claimed the spotlight in September. The artist's influence may have contributed to a unique auditory experience, shaping the musical landscape of the month.

**October**: October embraced the multifaceted sounds of Drake, coinciding with the release of his recent album "For All the Dogs." The album's release likely played a significant role in Drake's prominence throughout the month.

**November**: In November, Kanye West unsurprisingly emerged as the most played artist. The artist's emotionally charged tracks, such as "Ghost Town," provided a cathartic outlet during my challenges of the month.

### A Timeline of Minutes Played Throughout the Year
Concluding this overall analysis of my streaming history, I decided to create a timeline to visualize the minutes I spent immersed in music throughout the year.

```python
data['endTime'] = pd.to_datetime(data['endTime'])
data.set_index('endTime', inplace=True)
ms_played_per_day = data['msPlayed'].resample('D').sum()
plt.figure(figsize=(12, 6))
plt.plot(ms_played_per_day.index, ms_played_per_day.values, marker='o', linestyle='-', color='b')
plt.title('Minutes Played Throughout the Year')
plt.xlabel('Date')
plt.ylabel('Total msPlayed')
plt.grid(axis='both')
plt.show()
```
